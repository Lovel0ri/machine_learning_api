## 6.你能想到一个主动学习有用的示例吗，你将如何实施它？

主动学习是一种机器学习方法，其中模型在训练期间可以与用户进行交互，以获取更多的数据，从而改进其预测性能。以下是一个主动学习的示例：

假设你正在建立一个模型来预测房屋销售价格。你可以使用主动学习方法来改进模型的性能。首先，你可以从现有数据集中选择一小部分样本作为模型的初始训练数据。

然后，你可以利用模型对未标记的数据进行预测，并选择其中最具信息量的样本进行标记。例如，如果模型对某个房屋的预测误差最大，你可以选择标记该样本，并将其添加到训练数据中。这样做可以帮助模型更好地学习具有代表性的数据特征。

接下来，你可以重新训练模型，使用更新后的数据集。然后，你可以再次利用模型对未标记的数据进行预测，并选择具有最高信息量的样本进行标记。你可以继续重复这个过程，直到模型的性能达到满意的水平。

要实施这个方法，你需要一个初始的数据集、一个主动学习算法和一种方法来标记未标记的数据。你可以使用各种主动学习算法，如不确定性采样、噪声扰动采样或核心集采样。标记数据的方法可以是人工标记或自动标记。最后，你需要一个方法来评估模型的性能，以确定何时停止迭代过程。

总的来说，主动学习可以帮助你优化模型性能，并减少标记数据的成本。但是，它需要仔细的规划和实施，以确保有效性和可扩展性。

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 加载加州房价数据集
data = fetch_california_housing(as_frame=True).frame

# 将数据集拆分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-1], data.iloc[:,-1], test_size=0.2, random_state=42)

# 创建线性回归模型
model = LinearRegression()

# 在训练集上拟合模型
model.fit(X_train, y_train)

# 在测试集上评估模型
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("均方误差：", mse)
```

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 加载加州房价数据集
housing = fetch_california_housing()

# 将数据集分成已标记和未标记的子集
X_full, X_unlabeled, y_full, y_unlabeled = train_test_split(housing.data, housing.target, test_size=0.9, random_state=42)

# 将已标记子集进一步分成训练和验证集
X_train, X_valid, y_train, y_valid = train_test_split(X_full, y_full, test_size=0.2, random_state=42)

# 训练模型并在验证集上进行评估
model = LinearRegression()
model.fit(X_train, y_train)
y_pred_valid = model.predict(X_valid)
mse_valid = mean_squared_error(y_valid, y_pred_valid)
print("MSE on validation set:", mse_valid)

# 计算每个样本的预测方差
variances = np.var(model.predict(X_unlabeled), axis=0)

# 找到具有最大方差的样本的索引
top_uncertain_indices = np.argsort(variances)[::-1][:10]

# 将找到的样本添加到已标记集合中
X_full = np.concatenate([X_full, X_unlabeled[top_uncertain_indices]])
y_full = np.concatenate([y_full, y_unlabeled[top_uncertain_indices]])

# 移除已添加到已标记集合的样本
X_unlabeled = np.delete(X_unlabeled, top_uncertain_indices, axis=0)
y_unlabeled = np.delete(y_unlabeled, top_uncertain_indices)

# 训练一个新的模型并在测试集上进行评估
model2 = LinearRegression()
model2.fit(X_full, y_full)
y_pred_test = model2.predict(housing.data)
mse_test = mean_squared_error(housing.target, y_pred_test)
print("MSE on test set:", mse_test)
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 加载鸢尾花数据集
iris = load_iris()

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# 初始化标记集和未标记集
labeled_idx = []
unlabeled_idx = list(range(X_train.shape[0]))

# 随机选择 10 个样本加入标记集
labeled_idx.extend(np.random.choice(unlabeled_idx, size=10, replace=False))
unlabeled_idx = [i for i in unlabeled_idx if i not in labeled_idx]

# 初始化模型
model = LogisticRegression(max_iter=1000)

# 在标记集上训练模型
model.fit(X_train[labeled_idx], y_train[labeled_idx])

# 计算模型在测试集上的准确率
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy with initial labeled data: {:.2f}%".format(accuracy * 100))

# 开始主动学习循环，迭代 20 次
for i in range(10):
    # 计算每个样本的预测方差
    variances = np.var(model.predict_proba(X_train[unlabeled_idx]), axis=1)

    # 找到具有最大方差的样本的索引
    query_idx = unlabeled_idx[np.argmax(variances)]

    # 将这个样本加入标记集，并从未标记集中移除
    labeled_idx.append(query_idx)
    unlabeled_idx.remove(query_idx)

    # 在标记集上训练模型
    model.fit(X_train[labeled_idx], y_train[labeled_idx])

    # 计算模型在测试集上的准确率
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("Iteration {}: {:.2f}%".format(i + 1, accuracy * 100))
```

```python
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit

# 加载Olivetti人脸数据集
data = fetch_olivetti_faces()

# 将数据集拆分为训练集、验证集和测试集
X_train_val, X_test, y_train_val, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)

# 使用分层抽样将训练集和验证集进一步拆分为训练集和验证集
strat_split = StratifiedShuffleSplit(n_splits=1, test_size=0.25, random_state=42)
train_index, val_index = next(strat_split.split(X_train_val, y_train_val))
X_train, y_train = X_train_val[train_index], y_train_val[train_index]
X_val, y_val = X_train_val[val_index], y_train_val[val_index]
```
